# Cleaning

```python
country_list = df['Country'].unique()
metrics_list = df.keys()
metrics_list
```

```python
for country in country_list:
    df.loc[df['Country'] == country, metrics_list] = df.loc[df['Country'] == country,metrics_list].interpolate()
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
pd.options.display.float_format = '{:.2f}'.format
df.describe().T
```

# Histplots

```python
sns.set(font_scale=1.3)
data_numeric = df.select_dtypes(include='number')
column_numeric = data_numeric.columns.drop('Year')
```

```python
plt.figure(figsize=(24, 36))
plt.subplots_adjust(wspace=0.3, hspace=None)

i = 1
for col_name in column_numeric:
    plt.subplot(7,3,i)
    sns.histplot(data = df[col_name],  bins = 25)
    sns.rugplot(data = df[col_name], height=-0.02, clip_on=False)
    i +=1
```

# Boxplots

```python
plt.figure(figsize=(18,36))
plt.subplots_adjust(wspace=1, hspace=None)


i = 1
for col_name in column_numeric:
    plt.subplot(7,3,i)
    sns.boxplot(data=df, y=col_name)
    i +=1
```

# Correlation
### Pearson

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

> Початково було враховано демографічні змінні, структуру доходу та рівень смертності.
Зараз бачимо, що кількість населення зовсім не впливає. Значний вплив на очікувану тривалість життя мають: BMI (Average Body Mass Index of entire population), Income composition of resources (Human Development Index in terms of income composition of resources (index ranging from 0 to 1)), Number of years of Schooling(years), percentage expenditure (Expenditure on health as a percentage of Gross Domestic Product per capita(%)). Так само зворотній вплив мають: HIV/AIDS (Deaths per 1 000 live births HIV/AIDS (0-4 years)), thinness (Prevalence of thinness among children and adolescents for Age 5 to 9 and 10 to 19 (% ))

```python
plt.figure(figsize=(12,5))
plot = sns.scatterplot(data=df.groupby('Country').mean(), x='Life expectancy ',y='percentage expenditure',
                       hue='GDP', s=100)
plot.set(title='Life expectancy vs Percentage expenditure')
plt.show()
```

> Можемо побачити з графіка, що для зібльшення тривалості життя до 65+ країна має збільшити витрати на систему охорони здоров'я.

```python
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plot=sns.scatterplot(data=df.sort_values(by='Year'), y=' thinness  1-19 years', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Thinness  1-19 years of countries every year')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.legend([],[], frameon=False)
plt.subplot(2,1,2)
plot=sns.scatterplot(data=df.sort_values(by='Year'), y=' thinness 5-9 years', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Thinness 5-9 years of countries every year')
plt.legend([],[], frameon=False)
plt.show()
```

> З матриці кореляції ми знаємо, що коефіцієнт кореляції між очікуваною тривалістю життя та відсотком дітей з пониженою вагою рівний -0.46. На графіках наочно відображено цю негативну кореляцію.

```python
plt.figure(figsize=(12,5))
plot=sns.scatterplot(data=df.sort_values(by='Year'), y='', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Adult mortality of countries every year')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.legend([],[], frameon=False)
plt.show()
```

> Рівень дорослої смертності обернено пропорційно впливає на очікувану тривалість життя.

```python
plt.figure(figsize=(12,5))
plot=sns.scatterplot(data=df, y=' BMI ', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs BMI')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.legend([],[], frameon=False)
plt.show()
```

> Оскільки коефіцієнт кореліцяї між тривалістю життя та BMI (body mass index) рівний 0.6, можна стверджувати що спосіб життя та звички в їжі позитивно корелюють з тривалістю життя. Рівень споживання алкоголю також слабо позитивно корелює з тривалістю життя (0,39).

> Тривалість навчання дуже сильно позитивно корелює з тривалістю життя (0.8)

```python
plt.figure(figsize=(12,5))
plot=sns.scatterplot(data=df, y='Schooling', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Alcohol consumption')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.legend([],[], frameon=False)
plt.show()
```

> Тривалість життя майже ніяк не пов'язана з кількістю населення країни (-0.011).

```python
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plot=sns.scatterplot(data=df.sort_values(by='Year'), y='Diphtheria ', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Diphtheria immunisation of countries every year')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.legend([],[], frameon=False)
plt.subplot(2,1,2)
plot=sns.scatterplot(data=df.sort_values(by='Year'), y='Polio', x='Life expectancy ',
                     hue='Country')
plot.set(title='Life expectancy vs Polio immunisation of countries every year')
plt.legend([],[], frameon=False)
plt.show()
```

> Імунізація від дифтерії (0.44), поліолмієліту (0.42) та в меншій мірі гепатиту Б (0.25) позитивно корелюють з тривалістю життя.
